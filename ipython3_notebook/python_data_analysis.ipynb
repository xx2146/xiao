```python
import numpy as np
import time
```

```python
#Measuring time
strings = ['foo', 'foobar', 'baz', 'box', 'bill', 'sha', 'python', 'java'] * 10000

method_1 = [x for x in strings if x.startswith('foo') ]
method_2 = [x for x in strings if x[:3] == 'foo']
%time method_1
%time method_2
```

```python
%timeit [x for x in strings if x.startswith('foo') ]
```

```python
%timeit [x for x in strings if x[:3] == 'foo']
```
